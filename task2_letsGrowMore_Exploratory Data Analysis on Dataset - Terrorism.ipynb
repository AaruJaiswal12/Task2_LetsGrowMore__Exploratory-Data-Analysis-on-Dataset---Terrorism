# Exploratory Data Analysis on Dataset - Terrorism 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'globalterrorismdb_0718dist.csv',encoding="latin1")
```

# visualisation of data

```python
print("Maximally prefered type of Attacks:", df['attacktype1_txt'].value_counts().index[0])
```

```python
plt.subplots(figsize=(15,10))
sns.countplot('attacktype1_txt',data=df,palette='inferno',hue = df['targtype1_txt'])
plt.xticks(rotation=70)
plt.xlabel('Attack Type')
plt.title('Attacking Methods wiht Target as hue')
plt.show()
```

```python
print("Year with most number of Attacks:", df['iyear'].value_counts().index[0])
```

```python
plt.subplots(figsize=(15,8))
sns.countplot('iyear',data=df,palette='ocean',edgecolor=sns.color_palette('dark',7))
plt.xticks(rotation=90)
plt.xlabel('Years')
plt.title('Number Of Terrorist Attacks')
plt.show()
print("Maximum Attacks mostly happened during the year of ",df['iyear'].value_counts().index[0])
```

```python
plt.subplots(figsize=(15,8))
sns.countplot('imonth',data=df,palette='YlOrBr',edgecolor=sns.color_palette('dark',7))
plt.xticks(rotation=90)
plt.xlabel('Months')
plt.title('Number Of Terrorist Attacks')
plt.show()
print("Maximum Attacks mostly happened during the month of ",df['imonth'].value_counts().index[0])
```

```python
plt.subplots(figsize=(15,10))
sns.countplot('attacktype1_txt',data=df,palette='inferno')
plt.xticks(rotation=70)
plt.xlabel('Attack Type')
plt.title('Attacking Methods')
plt.show()
```

```python

df1 = df.sort_values(by='casualties',ascending=False)[:20]
df1
```

```python
heat =df1.pivot_table(index='country_txt',columns='iyear',values='casualties')
heat.fillna(0,inplace=True)
```

```python
heat.head()
```

```python

sns.heatmap(heat,cmap='summer')
plt.xlabel('Year')
plt.ylabel('Country')
plt.title('Heatmap for Attacked with respect to Country and Year')
```
